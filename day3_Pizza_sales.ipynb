```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('pizza_sales.csv')
df.head()
```

```python
unq_cat=df['pizza_category'].unique()
unq_size=df['pizza_size'].unique()

print("Pizza Categories:",unq_cat)
print("pizza size:",unq_size)
```

```python
#feature Engineering
#convert object type of  order_date and order_time to "date_time"data type
df['order_date']= pd.to_datetime(df['order_date'], format="mixed")
df['order_time']= pd.to_datetime(df['order_time'], format="mixed")
```

```python
df.info()
```

```python
df['order_month'] = df['order_date'].dt.month_name()
df['order_day'] = df['order_date'].dt.day_name()
df['order_hour'] = df['order_time'].dt.hour
```

```python
#sum of total_prize
total_revenue= df['total_price'].sum()

#sum of quantity sold
total_pizza_sold= df['quantity'].sum()

#count of unique order_id using nunique
unique_order= df['order_id'].nunique()

#average order value (AOV) =  total revenue divide total order
average_order_value= total_revenue / unique_order

#average pizza per order= total pizza sold divide total orders
average_pizza_per_order=total_pizza_sold / unique_order

print("Total Revenue:", total_revenue)
print("Total Pizza Sold:", total_pizza_sold)
print("Total Orders:", unique_order)
print("Average Order Value (AOV):", average_order_value)
print("Average Pizza per Order:", average_pizza_per_order)
```

```python
df['order_day'].unique()
```

```python
#custom category of day
weekend = ['Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
df['order_day']=pd.Categorical(categories=weekend,values=df['order_day'])
```

```python
day_sales = df.groupby('order_day',observed=False)['total_price'].sum().reset_index()
day_sales
```

```python
import matplotlib.pyplot as plt

# bar chart
plt.figure(figsize=(6,4))
plt.bar(day_sales['order_day'], day_sales['total_price'],
        color=['red','orange','blue','gray','purple'])

plt.xlabel('Day of week')
plt.ylabel('sales of weekdays')
plt.title('total sales of pizzas per weekdays')
plt.xticks(rotation=15)
plt.show()
```

```python
hour_sales= df.groupby('order_hour')['total_price'].sum().reset_index()
hour_sales
```

```python
plt.plot(hour_sales['order_hour'],hour_sales['total_price'],marker='o',c='green')
plt.xlabel('Order Hour')
plt.ylabel('Total sales per hour')
plt.grid(linestyle='--', color='black')
```

```python
monthly_sales= df.groupby('order_month')['total_price'].sum().reset_index()
monthly_sales
```

```python
sns.barplot(x=monthly_sales['order_month'],y=monthly_sales['total_price'],palette='viridis')
plt.xticks(rotation=25)
plt.xlabel('months')
plt.ylabel('total sales per month')
plt.title('monthly pizza sales')
```

```python
# % of sales by category

cat_sales = df.groupby('pizza_category')['total_price'].sum().reset_index()

# calculate percentage of total revenue
cat_sales['price_pct'] = (cat_sales['total_price'] / total_revenue) * 100

# format percentage
cat_sales['price_pct'] = cat_sales['price_pct'].map('{:.2f}'.format)

cat_sales
```

```python
  
plt.bar(cat_sales['pizza_category'], cat_sales['price_pct'])
plt.xlabel('pizza category')
plt.ylabel('total sales percentage')
plt.title('percentage  total pizza sold by pizza category')
```

```python
size_sales = df.groupby('pizza_size')['total_price'].sum().reset_index()

# Optional: calculate percentage of total sales
size_sales['price_pct'] = (size_sales['total_price'] / size_sales['total_price'].sum()) * 100
size_sales['price_pct'] = size_sales['price_pct'].round(2)

# Display the table
size_sales
```

```python
size_sales = df.groupby('pizza_size')['total_price'].sum().reset_index()


size_sales['price_pct'] = (size_sales['total_price'] / size_sales['total_price'].sum()) * 100
size_sales['price_pct'] = size_sales['price_pct'].round(2)


plt.figure(figsize=(7,5))
plt.bar(size_sales['pizza_size'], size_sales['price_pct'], color='green')

plt.xlabel('Pizza Size')
plt.ylabel('Percentage of Total Sales (%)')
plt.title('Pizza Sales by Size')

# Optional: show percentage on top of bars
for index, value in enumerate(size_sales['price_pct']):
    plt.text(index, value + 0.5, str(value) + '%', ha='center')

plt.show()
```

```python
p_cat_sales = df.groupby('pizza_category')['total_price'].sum().reset_index()


p_cat_sales['pct_of_total'] = (p_cat_sales['total_price'] / p_cat_sales['total_price'].sum()) * 100
p_cat_sales['pct_of_total'] = p_cat_sales['pct_of_total'].round(2)

plt.figure(figsize=(8,5))
plt.bar(p_cat_sales['pizza_category'], p_cat_sales['total_price'], color='purple')

# Labels and title
plt.xlabel('Pizza Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Pizza Category')


plt.xticks(rotation=45)
plt.show()
```

```python
#Top  5 best selling pizzas 
#  A horizontal bar chart showing pizzas with the highest sales(by revenue,orders or quality)
#Support promotional and menu strategy

best_5_pizza = df.groupby('pizza_name_id')['total_price'].sum().sort_values(ascending=False).reset_index()
best_5_pizza
```

```python
#Top  5 best selling pizzas 
#  A horizontal bar chart showing pizzas with the highest sales(by revenue,orders or quality)
#Support promotional and menu strategy

best_5_pizza = df.groupby('pizza_name_id')['total_price'].sum().sort_values(
                                                                ascending=False
                                                                ).head(5).reset_index()
best_5_pizza

sns.barplot(data=best_5_pizza,x='pizza_name_id',y='total_price')
```

```python
# Top 5 best-selling pizzas by revenue
best_5_pizza = df.groupby('pizza_name_id')['total_price'].sum().sort_values(
    ascending=False
).head(5).reset_index()

# Horizontal bar chart
sns.barplot(data=best_5_pizza, x='total_price', y='pizza_name_id', palette='coolwarm')

plt.xlabel('Total Sales (Revenue)')
plt.ylabel('Pizza Name')
plt.title('Top 5 Best-Selling Pizzas')
plt.show()
```

```python
#Bottom 5 Least-Selling Pizzas
#A horizontal bar chart of pizzas with the lowest sales.
# Identifies products for improvement or possible removal from the menu.
# Bottom 5 least-selling pizzas by revenue
bottom_5_pizza = df.groupby('pizza_name_id')['total_price'].sum().sort_values(
    ascending=True
).head(5).reset_index()

# Horizontal bar chart
sns.barplot(data=bottom_5_pizza, x='total_price', y='pizza_name_id', palette='magma')

plt.xlabel('Total Sales (Revenue)')
plt.ylabel('Pizza Name')
plt.title('Bottom 5 Least-Selling Pizzas')
plt.show()
```
