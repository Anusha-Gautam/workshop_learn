```python
hr=pd.read_csv('HR_Dataset Refresh.csv')
hr.head()
```

```python
#Feature Engineering
'''feature Engineering is a concept of extracting some required data from original dataset for further data analysis to be done
For example: extracting year, month and day from date column.
'''
```

```python
hr['DateofHire'].dtype
```

```python
#convert dateofhire to datetime datatype
hr['DateofHire']=pd.to_datetime(hr['DateofHire'])
```

```python
#extract year from date
hr['Hire_year']= hr['DateofHire'].dt.year
```

```python
hr['Hire_Month_Name'] =hr['DateofHire'].dt.month_name()
hr['Hire_Day_Name'] = hr['DateofHire'].dt.day_name()
```

```python
hr
```

```python
#strip()-> Remove unwanted right and left spaces from data
#Remove white space from columns
```

```python

hr['Department'].unique()
```

```python
hr['Gender']=hr['Gender'].str.strip()
```

```python
#data analysis
#groupby and aggregate functions are used while data analysis using python pandas.
#GroupBy->categorical record
#aggregate->Numeric
```

```python
#Find total salary of employee associate with each departments.
''' 
syntax:
      emp_sal=dataframe.groupby(['columnname1','columnName2'])['Aggcolumn'].sum()

      single Group by
      emp_sal=dataframe.groupby('columname2)
'''
```

```python


emp_sal = hr.groupby('Department')['Salary'].sum().reset_index()
(emp_sal)
```

```python
emp_dep = hr.groupby('Department')['EmpID'].count().reset_index()
(emp_dep)
```

```python
emp_dep = hr.groupby('Department')['EmpID'].count().reset_index().rename(
    columns={'EmpID': 'Total_Number_Of_Employees'}
)

emp_dep
```

```python
emp_dep = hr.groupby('Department').agg(
    total_no_employee=('EmpID', 'count')
).reset_index()

print(emp_dep)
```

```python
emp_dep = hr.groupby('Department').agg(
    Total_No_Employee = ('EmpID', 'count'),
    Total_Salary =('Salary','sum'),
    Avg_Salary = ('Salary','mean')
).reset_index()
emp_dep
```

```python
#find details of departments which has less than 40 employees
total_emp = emp_dep['Total_No_Employee']<40
emp_dep[total_emp]
```

```python
emp_dep['Total_Salary'] = emp_dep['Total_Salary'].map("Rs.{:,.2f}".format)
```

```python
emp_dep
```

```python
emp_dep['Avg_Salary'] = emp_dep['Avg_Salary'].map("Rs.{:,.2f}".format)

emp_dep
```

```python
fruits=["Apple", "Banana","Coconut", "Papaya","mango"]
sales=[200,150,300,400,250]
revenue=[100,50,150,250,140]
```

```python
plt.figure(figsize=(6,4))
plt.plot(fruits, sales, c='red', marker='o',alpha=0.7,label='Sales')
plt.plot(fruits, revenue, c='blue', marker='d',alpha=0.7,ls='dashed',label='Revenue')
plt.xlabel('Fruits')
plt.ylabel('Total Sales of Fruits')
plt.title('Fruit Sales Overview')
plt.legend()
#extensions of figure -> .png, .jpg ,.jpeg
plt.savefig('sales.jpg')
```
