```python
try:
    a=int(input("enter  a number:"))
    b=int(input("enter a number:"))
    div=a/b 
    print(div)
except:
    print('Error:A number cannot be divisible.')
```

```python
try:
    a=int(input("enter  a number:"))
    b=int(input("enter a number:"))
    div=a/b 
except ZeroDivisionError:
    print('Error:A number cannot be divisible.')
except ValueError:
    print("Error:only target integer value id acceptable.")
else:#all print value will be inside else block
    print(div)
finally:
    print("division is completed.")
```
